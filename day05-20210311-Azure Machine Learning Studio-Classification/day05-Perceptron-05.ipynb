```python
from sklearn import datasets
iris = datasets.load_iris()
iris.keys()
```

```python
print(iris.DESCR)
```

```python
# 花的數據
'''
        - sepal length in cm 花萼 長度
        - sepal width in cm  花萼 寬度
        - petal length in cm 花瓣 長度
        - petal width in cm  花瓣 寬度
'''    
#
iris.data
```

```python
# 花的種類
'''
                - Iris-Setosa      0
                - Iris-Versicolour 1
                - Iris-Virginica   2
'''
iris.target
```

```python
# 認識 特徵 features (花萼長度 花萼寬度 花瓣長度 花瓣寬度)
# 認識 標籤 labels   (Iris-Setosa-0, Iris-Versicolour-1, Iris-Virginica-2)
# 資料集 csv
'''
slength swidth plength pwidth class

'''
import pandas as pd
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns=['slength','swidth','plength','pwidth'])
df['class'] = y
df[-3:]
# 輸出到 csv 不要 index 
df.to_csv('iris.csv', index = False)
```

```python
import pandas as pd
lemonade = pd.read_csv('Lemonade.csv')
lemonade[:3]
```

```python
# 針對 Temperature, Rainfall , Flyers, Price 與 Sales 分析圖
import matplotlib.pyplot as plt
for c in lemonade.columns[2:-2]:    
    #print(c)
    plt.title('{} vs Sales'.format(c))
    plt.scatter(lemonade[c], lemonade.Sales)
    plt.show()
```

```python
# 針對 Temperature, Rainfall , Flyers, Price 與 Sales 分析圖
# 取對數 np.log 後的 分析圖
import numpy as np
import matplotlib.pyplot as plt
for c in lemonade.columns[2:-2]:    
    #print(c)
    plt.title('{} vs Sales'.format(c))
    plt.scatter(np.log(lemonade[c]), np.log(lemonade.Sales) )
    plt.show()
```

```python
# 請劃出 溫度與銷售
import matplotlib.pyplot as plt
plt.scatter(lemonade.Temperature, lemonade.Sales)
plt.show()
```

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html
# 利用亂數產生樣本 根據中心點產生
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=10000, centers=4, n_features=2, random_state=0)
                # 所有資料列     種類       欄位數量   
```

```python
# 繪畫出 scatter 數據 分成三種顏色
import numpy as np
import matplotlib.pyplot as plt
colors = ['r', 'b', 'g', 'm']
plt.figure(figsize=(12,4))
for i in np.unique(y):
    #print( X[y==i], y[y==i] )
    #print( X[y==i][:, 0], X[y==i][:,1] )
    plt.scatter(X[y==i][:, 0], X[y==i][:, 1] , cmap = colors[i])
plt.show()
```

```python
# 使用 KMEANS 分群
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
model.fit(X)
```

```python
y_pred = model.predict(X)
```

```python
import numpy as np
#  注意 對應群組 代號 不會一樣
y_pred[:5], y[:5]
```

```python
# 繪畫出 scatter 數據 分成三種顏色
import numpy as np
import matplotlib.pyplot as plt
colors = ['r', 'b', 'g', 'm', 'o']
plt.figure(figsize=(12,4))
for i in np.unique(y_pred):

    plt.scatter(X[y_pred==i][:, 0], X[y_pred==i][:, 1] , cmap = colors[i])
plt.show()
```

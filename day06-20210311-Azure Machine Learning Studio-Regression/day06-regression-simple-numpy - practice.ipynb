```python
import pandas as pd
df = pd.read_csv('lemonade.csv')
df[:3]
```

```python
# 畫圖觀察特徵(features)與標籤(label = Sales)的關係
import matplotlib.pyplot as plt
for f in ['Temperature', 'Rainfall', 'Flyers']:
    plt.figure(figsize=(12,3))
    plt.title('The relationship between {} and Sales'.format(f))
    plt.scatter(df[f], df.Sales, color ='red')
    plt.grid()
    plt.show()
```

```python
# 加上對數
# 畫圖觀察特徵(features)與標籤(label = Sales)的關係
import numpy as np
import matplotlib.pyplot as plt
for f in ['Temperature', 'Rainfall', 'Flyers']:
    plt.figure(figsize=(12,3))
    plt.title('The relationship between {} and Sales'.format(f))
    plt.scatter(np.log(df[f]), np.log(df.Sales), color ='blue')
    plt.grid()
    plt.show()
```

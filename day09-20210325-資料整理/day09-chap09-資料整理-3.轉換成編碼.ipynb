```python
code = {'A':0, 'B':1, 'C':2}
```

```python
import pandas as pd
df = pd.read_csv('datae.txt')
df
```

```python
df['LEVELY'] = [code[x] for x in df.GRADE]
df
```

```python
# 補充使用LOOP
df['LEVELX'] = 0
for i, v in enumerate(df.GRADE):
    print(i, v, code[v])
    df.loc[df.index==i, 'LEVELX'] = code[v]

df
```

```python
# label encoding 
df.GRADE.apply(lambda x : code[x])
df['LEVEL'] = df.GRADE.apply(lambda x : code[x])
#df.to_excel('dfe.xlsx', index = False)
df
```

```python
# 多出多少欄位 要從unique values
df.GRADE.unique()
```

```python
for c in df.GRADE.unique():
    df['LEVEL'+c] = 0
df    
```

```python
# 補充使用LOOP
for i, v in enumerate(df.GRADE):
    print(i, v, code[v])
    df.loc[df.index==i, 'LEVEL'+v] = 1 # 針對位置去變更數據
df
```

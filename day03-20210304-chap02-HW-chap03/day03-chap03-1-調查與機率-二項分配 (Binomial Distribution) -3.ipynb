```python
# https://zh.wikipedia.org/wiki/%E4%BA%8C%E9%A0%85%E5%BC%8F%E5%88%86%E5%B8%83
import math
import matplotlib.pyplot as plt
import numpy as np
#print(dir(math))
#math.factorial(3)
def c(n, k):
    return( math.factorial(n) / (math.factorial(k) * math.factorial(n-k)) )

def f(k, n, p):
    return( c(n, k) * p**k * (1-p)**(n-k) )


plt.figure(figsize = (12,4))
# p=0.5 , n=20
x = np.arange(0, 21, 1)
y = [f(k=xx, n=20, p=0.5 ) for xx in x]
print(y)
plt.scatter(x, y, color='blue')


# p=0.7 , n=20
x = np.arange(0, 21, 1)
y = [f(k=xx, n=20, p=0.7 ) for xx in x]
plt.scatter(x, y, color='green')

# p=0.5 , n=40
x = np.arange(0, 41, 1)
y = [f(k=xx, n=40, p=0.5 ) for xx in x]
plt.scatter(x, y, color='red')

plt.grid()
plt.show()
```

```python
# https://zh.wikipedia.org/wiki/%E4%BA%8C%E9%A0%85%E5%BC%8F%E5%88%86%E5%B8%83
import math
import matplotlib.pyplot as plt
import numpy as np
#print(dir(math))
#math.factorial(3)
def c(n, k):
    return( math.factorial(n) / (math.factorial(k) * math.factorial(n-k)) )

def f(k, n, p):
    return( c(n, k) * p**k * (1-p)**(n-k) )


plt.figure(figsize = (12,4))
# p=0.5 , n=20
x = np.arange(0, 21, 1)
y = np.cumsum([f(k=xx, n=20, p=0.5 ) for xx in x])
plt.scatter(x, y, color='blue')


# p=0.7 , n=20
x = np.arange(0, 21, 1)
y = np.cumsum([f(k=xx, n=20, p=0.7 ) for xx in x])
plt.scatter(x, y, color='green')

# p=0.5 , n=40
x = np.arange(0, 41, 1)
y = np.cumsum([f(k=xx, n=40, p=0.5 ) for xx in x])
plt.scatter(x, y, color='red')

plt.grid()
plt.show()
```

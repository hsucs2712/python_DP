```python
#某公司市場調查後發現，消費者有25%機率喜歡這家公司產品
#請問重複訪問 30個人  100次，大家喜歡公司產品的機率以圖表方式顯示

import numpy.random as nr
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
# 一個人問一次 喜歡機率25% 連續問三十人
results = [np.mean(nr.binomial(1, 0.25, 30)) for i in range(100)]
#results
# [np.mean(nr.binomial(1, 0.25, 30)) for i in range(100)]
sample_mean = np.mean(results)
plt.hist(results)
plt.vlines(0.25, 0.0, 28.0, color = 'red')
plt.vlines(sample_mean, 0.0, 28.0, color = 'black')
plt.xlabel('Results')
plt.ylabel('Frequency')
plt.title('Histogram of results')
plt.show()
```

```python
# https://zh.wikipedia.org/zh-tw/%E5%8D%9C%E7%93%A6%E6%9D%BE%E5%88%86%E5%B8%83
import math
import numpy as np
import matplotlib.pyplot as plt
def f(lbd, k):
    return( (np.exp(-1*lbd) * lbd**k  ) / (math.factorial(k))   )
plt.figure(figsize = (12,4))
x = range(21)
y = np.cumsum([f(lbd=10, k=xx) for xx in x])
plt.scatter(x, y, color='blue')

x = range(21)
y = np.cumsum([f(lbd=4, k=xx) for xx in x])
plt.scatter(x, y, color='purple')

x = range(21)
y = np.cumsum([f(lbd=1, k=xx) for xx in x])
plt.scatter(x, y, color='yellow')

plt.grid()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('Lemonade.csv')
data[:3]
```

```python
X, y = data.iloc[:, 2:-1].values, data.iloc[:, -1].values
X, y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# 計算誤差的部分
import pandas as pd
import numpy as np
dfTest = pd.DataFrame(X_test, columns = ['Temperature', 'Rainfall', 'Flyers', 'Price'])
dfTest['Sales'] = y_test
dfTest['Pre_Sales'] = model.predict(X_test)
dfTest['Diff_Sales'] = np.abs(dfTest['Sales'] - dfTest['Pre_Sales'])
dfTest['Diff_Sales_%'] = np.round(np.abs(dfTest['Sales'] - dfTest['Pre_Sales']) / dfTest['Sales'] * 100, 2)
dfTest['Diff_Sales_Sqr'] = np.power(dfTest['Sales'] - dfTest['Pre_Sales'], 2)
print( ( dfTest['Diff_Sales_Sqr'].sum() / len(dfTest['Diff_Sales_Sqr']) ) ** 0.5 ) # RMSE 越小越好 SUM(差距平方)/N 再開根號
dfTest.sort_values(by='Diff_Sales_%', ascending = False)[:10]
```
